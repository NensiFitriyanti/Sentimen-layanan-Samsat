```python
3.	Analisis Frekuensi Kata Dominan
Untuk mengidentifikasi tema atau topik yang paling sering dibicarakan masyarakat, dilakukan analisis frekuensi kata dari komentar yang telah distem menggunakan algoritma dari pustaka Sastrawi. Setelah melalui tahapan pembersihan, tokenisasi, stopword removal, dan stemming, semua kata dari komentar dihimpun dan dihitung frekuensinya menggunakan nltk.FreqDist.
Hasil analisis menunjukkan bahwa kata-kata seperti layan, cepat, samsat, ramah, dan antre termasuk kata yang paling sering muncul. Kata-kata tersebut mencerminkan fokus utama opini masyarakat, baik dalam bentuk apresiasi terhadap pelayanan cepat dan ramah, maupun keluhan terhadap waktu tunggu dan antrean panjang. Kata-kata yang muncul secara konsisten dapat dijadikan indikator awal untuk mengidentifikasi aspek layanan yang perlu dipertahankan dan yang perlu ditingkatkan.
import pandas as pd
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import nltk
import seaborn as sns # Added seaborn import for the bar plot

# Download necessary NLTK data
nltk.download('punkt_tab', quiet=True) # Download punkt_tab as suggested by the error
nltk.download('stopwords', quiet=True)


# Assuming df is already loaded with 'Cleaned_Komentar' column
# Combine all cleaned comments into a single string
all_comments_text = " ".join(df['Cleaned_Komentar'].dropna().astype(str))

# Tokenize the text
words = word_tokenize(all_comments_text)

# Remove stopwords (optional but recommended for frequency analysis)
# You might need to add Indonesian stopwords if available, or create a custom list
stopwords_english = set(nltk.corpus.stopwords.words('english'))
# Example of custom Indonesian stopwords - you might want to expand this
stopwords_indonesian = set(['yang', 'dan', 'di', 'dari', 'dengan', 'untuk', 'itu', 'ini', 'tidak', 'ada', 'akan', 'bisa', 'sudah', 'sangat', 'lebih', 'kurang', 'saat', 'tahun', 'bulan', 'hari', 'jam', 'kalau', 'kalau', 'atau', 'tapi', 'juga', 'pada', 'ke', 'dari', 'oleh', 'karena', 'sehingga', 'agar', 'supaya', 'meskipun', 'walaupun', 'yaitu', 'yakni', 'ialah', 'adalah', 'yaitu', 'yakni', 'ialah', 'adalah', 'yaitu', 'yakni', 'ialah', 'adalah'])

filtered_words = [word.lower() for word in words if word.lower() not in stopwords_english and word.lower() not in stopwords_indonesian and word.isalpha()]


# Calculate frequency distribution
fdist = FreqDist(filtered_words)

# Get the top 20 most frequent words
top_words = fdist.most_common(20)

# Convert to DataFrame for easier plotting
df_top_words = pd.DataFrame(top_words, columns=['Word', 'Frequency'])

# Display the top words and their frequencies
print("20 Kata Paling Sering Muncul:")
display(df_top_words)

# Create a bar chart of the top 20 words
plt.figure(figsize=(12, 7))
sns.barplot(x='Frequency', y='Word', data=df_top_words, palette='viridis')
plt.title('20 Kata Paling Sering Muncul dalam Komentar')
plt.xlabel('Frekuensi')
plt.ylabel('Kata')
plt.tight_layout()
plt.show()
```

```python
import nltk
from nltk.tokenize import word_tokenize
import pandas as pd

# Download necessary NLTK data (if not already downloaded)
try:
    nltk.data.find('tokenizers/punkt')
except nltk.downloader.DownloadError:
    nltk.download('punkt', quiet=True)
try:
    nltk.data.find('corpora/stopwords')
except nltk.downloader.DownloadError:
    nltk.download('stopwords', quiet=True)

# Assuming df_book1 is already loaded and 'Unnamed: 4' is the comment column

# Extract the comment column and drop missing values
comments_text = df_book1['Unnamed: 4'].dropna().astype(str)

# Remove the header row 'Isi Komentar' if it's present in the data
comments_text = comments_text[comments_text != 'Isi Komentar']

# Combine all cleaned comments into a single string for frequency analysis
all_comments_text_new = " ".join(comments_text)

# Tokenize the text
words_new = word_tokenize(all_comments_text_new)

# Load Indonesian stopwords
stopwords_indonesian = set(nltk.corpus.stopwords.words('indonesian'))

# Add some custom stopwords if needed (e.g., very frequent words that are not informative)
custom_stopwords = set(['yg', 'saya', 'mau', 'min', 'ya', 'nya', 'kalo', 'ga', 'deh', 'pak', 'bu', 'kak', 'ko', 'ni', 'udah', 'kali', 'aja', 'bisa', 'masih']) # Add custom stopwords here
all_stopwords = stopwords_indonesian.union(custom_stopwords)

# Remove stopwords and non-alphabetic tokens
filtered_words_new = [word.lower() for word in words_new if word.lower() not in all_stopwords and word.isalpha()]

print(f"Jumlah total kata setelah preprocessing: {len(filtered_words_new)}")
print("Beberapa kata pertama setelah preprocessing:")
# Display the first 20 words vertically (as a list)
display(filtered_words_new[:20])
```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize

# Download necessary NLTK data (if not already downloaded)
try:
    nltk.data.find('tokenizers/punkt')
except nltk.downloader.DownloadError:
    nltk.download('punkt', quiet=True)
try:
    nltk.data.find('corpora/stopwords')
except nltk.downloader.DownloadError:
    nltk.download('stopwords', quiet=True)

# Assuming df_book1 is already loaded and 'Unnamed: 4' is the comment column

# Extract the comment column and drop missing values
comments_text_for_table = df_book1['Unnamed: 4'].dropna().astype(str)

# Remove the header row 'Isi Komentar' if it's present in the data
comments_text_for_table = comments_text_for_table[comments_text_for_table != 'Isi Komentar'].reset_index(drop=True)

# Load Indonesian stopwords
stopwords_indonesian = set(nltk.corpus.stopwords.words('indonesian'))

# Add some custom stopwords if needed (e.g., very frequent words that are not informative)
custom_stopwords = set(['yg', 'saya', 'mau', 'min', 'ya', 'nya', 'kalo', 'ga', 'deh', 'pak', 'bu', 'kak', 'ko', 'ni', 'udah', 'kali', 'aja', 'bisa', 'masih']) # Add custom stopwords here
all_stopwords = stopwords_indonesian.union(custom_stopwords)

# Apply tokenization and stopword removal to each comment
tokenized_comments = [word_tokenize(comment.lower()) for comment in comments_text_for_table]
filtered_comments = [[word for word in tokens if word not in all_stopwords and word.isalpha()] for tokens in tokenized_comments]

# Create a DataFrame for visualization
# We will display only the first few examples
num_examples = 5
table_data = {
    'Komentar Asli': comments_text_for_table.head(num_examples),
    'Setelah Tokenisasi': [tokenized_comments[i] for i in range(num_examples)],
    'Setelah Stopword Removal': [filtered_comments[i] for i in range(num_examples)]
}

df_preprocessing_example = pd.DataFrame(table_data)

print(f"Contoh {num_examples} Komentar Setelah Tahap Pra-pemrosesan:")
display(df_preprocessing_example)
```
